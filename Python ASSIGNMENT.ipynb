1. **Custom Exception Handling:**

*o Define a custom exception class InvalidInputError to handle invalid inputs.*

```python
class InvalidInputError(Exception):
    """Custom exception for invalid inputs."""
    pass
try:
  invalid=int(input())
  if invalid > 30 or invalid < 10:
    raise InvalidInputError

except InvalidInputError:

  print("Invalid Input")
else:
  print("Valid Input")
```

2. **Exception Blocks:**


*o Use try, except, else, and finally blocks to handle exceptions properly.*

```python
try:
  a=int(input('enter the first number:'))
  b=int(input('enter the second number:'))
  c=a/b
  print(c)
except ZeroDivisionError:
  print('*Zero Division Error*')
except ValueError:
  print('*Value Error*')
```
